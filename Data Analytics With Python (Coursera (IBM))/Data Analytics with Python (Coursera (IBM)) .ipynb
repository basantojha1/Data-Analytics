```python
# Identifying and handeling missing data
# Replacing ? with NaN
df.replace('?', np.NaN, inplace=True)
df.head()
```

```python
# Evaluating for missing data
df.isnull().sum()
```

```python
# As we can see in the above data normalized-losses has the maximum missing values
```

```python
# Counting Missing values in each column
missing_data = df.isnull()

for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print('')
```

```python
# Dealing with missing values
'''
Drop the whole column
drop the whole row

replace it by mean
replace it by frequency
replace it based on other functions
'''
```

```python
# Checking normalized losses
df['normalized-losses']
```

```python
# For normalized-losses
# Calculating the mean value for the normalize-losses
avg_norm_loss = df['normalized-losses'].astype('float').mean(axis=0)
print('Average of Normalized Losses:', avg_norm_loss)
```

```python
# Replacing NaN Value in normalized-losses column
df['normalized-losses'].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
# Checking for null values
df['normalized-losses'].isnull().sum()
```

```python
# For bore
# Calculating the mean value for bore
avg_bore = df['bore'].astype('float').mean(axis=0)
print('Average of bore:', avg_bore)
```

```python
# Replacing the value
df['bore'].replace(np.nan, avg_bore, inplace=True)
```

```python
df['bore'].isnull().sum()
```

```python
# For stroke
# calculating the mean value for stroke
avg_stroke = df['stroke'].astype('float').mean(axis=0)
print('Average of stroke:', avg_stroke)
```

```python
# replacing null values with mean
df['stroke'].replace(np.nan, avg_stroke, inplace=True)
```

```python
df['stroke'].isnull().sum()
```

```python
# For horsepower
# calculating the mean value for horsepower
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print('Average horsepower:', avg_horsepower)
```

```python
# replacing null values with the mean value
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
df['horsepower'].isnull().sum()
```

```python
# For peak-rpm
avg_peak_rpm = df['peak-rpm'].astype('float').mean(axis=0)
print('Average Peak RPM:', avg_peak_rpm)
```

```python
# Replacing the null values with average peak rpm
df['peak-rpm'].replace(np.nan, avg_peak_rpm, inplace=True)
```

```python
df['peak-rpm'].isnull().sum()
```

```python
# For number of doors
df['num-of-doors'].value_counts()
```

```python
# we can also use .idxmax() for checking the most common type automatically
df['num-of-doors'].value_counts().idxmax()
```

```python
# Replacing the null values with four
df['num-of-doors'].replace(np.nan, 'four', inplace=True)
```

```python
# Checking again if there is any null values remaing
df.isnull().sum()
```

```python
df.head()
```

```python
# Changing data formate
df.dtypes
```

```python
# Converting datatypes in the correct formate
df[['bore', 'stroke']] = df[['bore', 'stroke']].astype('float')
df[['normalized-losses']] = df[['normalized-losses']].astype('int')
df[['price']] = df[['price']].astype('float')
df[['peak-rpm']] = df[['peak-rpm']].astype('float')
```

```python
df.dtypes
```

```python
# Standardization
df.head()
```

```python
# Converting mpg to L/100km by using mathematical operation
# L/100km = 235 divided by mpg
df['city-L/100km'] = 235/df['city-mpg']
df.head()
```

```python
# Converting highway-mpg to highway-L/100km
df['highway-mpg'] = 235/df['highway-mpg']
# Remaning the column name
df.rename(columns = {'highway-mpg':'highway-L/100km'}, inplace=True)
df.head()
```

```python
# Normalization
# Nornalization approach: Replace orginal value by (Orginal value)/(Maximum value)
```

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
df[['length', 'width', 'height']].head()
```

```python
# We have Normalized Lenght breadth and height in range of [0,1]
```

# Example of Binning data in Pandas

```python
# First step: Convert the data into correct formate
df['horsepower'] = df['horsepower'].astype(int, copy=True)
```

```python
# Let us plot a histogram of horsepower to see the distribution
plt.hist(df['horsepower'])

# Setting labels and title
plt.xlabel('Horsepower')
plt.ylabel('Count')
plt.title('Horsepower Bins')
```

```python
# Creating bin array
bins = np.linspace(min(df['horsepower']), max(df['horsepower']), 4)
bins
```

```python
# We set group names
group_names = ['Low', 'Medium', 'High']
```

```python
# Applying cut function to determine where each value belongs
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True)
df[['horsepower', 'horsepower-binned']].head(15)
```

```python
# Let us check the mumber of vehicles in each bins
df['horsepower-binned'].value_counts()
```

```python
# Plotting a barchart
plt.bar(group_names, df['horsepower-binned'].value_counts())

# Setting label and title
plt.xlabel('Horsepower Level')
plt.ylabel('Count')
plt.title('Horsepower Binned')
```

```python
# Drawing histogram for horsepower with bins 3
plt.hist(df['horsepower'], bins = 3)
```
